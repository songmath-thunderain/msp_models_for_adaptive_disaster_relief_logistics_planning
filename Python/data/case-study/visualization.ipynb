```python
import folium
import pgeocode
```

```python
nomi = pgeocode.Nominatim('us')
```

```python
# Create a map centered around South Carolina
map_sc = folium.Map(location=[33.8361, -81.1637], zoom_start=8)

# Add markers for each data point
data = [
    {'SN': 1, 'Group': 'C1', 'Capacity': 14854, 'ZIP': 29418},
    {'SN': 2, 'Group': 'C2', 'Capacity': 8852, 'ZIP': 29054},
    {'SN': 3, 'Group': 'C3', 'Capacity': 3846, 'ZIP': 29135},
    {'SN': 4, 'Group': 'C4', 'Capacity': 2260, 'ZIP': 29420},
    {'SN': 5, 'Group': 'N1', 'Capacity': 13780, 'ZIP': 29554},
    {'SN': 6, 'Group': 'N2', 'Capacity': 3050, 'ZIP': 29550},
    {'SN': 7, 'Group': 'N3', 'Capacity': 5168, 'ZIP': 29010},
    {'SN': 8, 'Group': 'N4', 'Capacity': 5324, 'ZIP': 29501},
    {'SN': 9, 'Group': 'S1', 'Capacity': 5019, 'ZIP': 29803},
    {'SN': 10, 'Group': 'S2', 'Capacity': 2026, 'ZIP': 29817},
    {'SN': 11, 'Group': 'S3', 'Capacity': 350, 'ZIP': 29853},
]

for item in data:
    lat = nomi.query_postal_code(item['ZIP'])["latitude"]  # Function to get latitude based on ZIP code
    lon = nomi.query_postal_code(item['ZIP'])["longitude"]  # Function to get longitude based on ZIP code
    
    marker = folium.Marker(
        location=[lat, lon],
        popup=f"Group: {item['Group']}<br>Capacity: {item['Capacity']}",
        tooltip=item['SN']
    )
    marker.add_to(map_sc)

# Display the map
map_sc
```
