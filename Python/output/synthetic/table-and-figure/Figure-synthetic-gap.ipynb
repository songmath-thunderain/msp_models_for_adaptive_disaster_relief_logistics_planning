```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def createFigure(inputname, outputname):
    # Load the data from the CSV file
    #file_path = 'for-Figure-synthetic-gap-1.csv'
    data = pd.read_csv(inputname)

    # Calculate the average percentage gap for each value of nu
    average_gaps = data.groupby('nu')[['MSP', 'Decision-tree', 'WS', 'rh-2SSP', 's-2SSP']].mean()
    print(average_gaps)
    # Plot the results using a bar chart
    average_gaps.plot(kind='bar', figsize=(10, 6))
    plt.title(r"Average Percentage Gap for Different Values of $\nu$")
    plt.xlabel(r"$\nu$")
    plt.ylabel('Average Percentage Gap (%)')
    plt.xticks(rotation=0)
    plt.legend(title='Approach',loc="upper right")
    plt.grid(axis='y')
    # Save the figure to a file
    #plt.savefig('average_percentage_gap1.png')
    plt.savefig(outputname)
    plt.show()
```

```python
#createFigure('for-Figure-synthetic-gap-1.csv','average_percentage_gap1.png')
#createFigure('for-Figure-synthetic-gap-2.csv','average_percentage_gap2.png')
createFigure('for-Figure-synthetic-gap-3.csv','average_percentage_gap3.png')
#createFigure('for-Figure-synthetic-gap-4.csv','average_percentage_gap4.png')
```
