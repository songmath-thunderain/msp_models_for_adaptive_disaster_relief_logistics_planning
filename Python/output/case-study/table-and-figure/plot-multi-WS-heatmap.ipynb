```python
# Function to create heatmap data from dictionary
def create_heatmap_data(data, max_time, max_intensity):
    heatmap_data = np.full((max_intensity + 1, max_time + 1), np.nan)
    for (time, intensity), value in data.items():
        if value != -1:
            heatmap_data[intensity, time] = value
    return heatmap_data
```
