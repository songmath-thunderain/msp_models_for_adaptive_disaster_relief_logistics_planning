```julia
using CSV, DataFrames, Plots, Plots.PlotMeasures, LaTeXStrings, XLSX#, GR
```

```julia
xf = XLSX.readxlsx("procurement-avgAmount-case-study.xlsx")
sh = xf["all"] # get a reference to a Worksheet
dt = sh["A1:I3"];
rhos = [1,2,3,4,5,6,7,8,9];
vals = convert.(Float64,dt);
vals = transpose(vals);
policies = [L"\nu = 0.05", L"\nu = 0.5", L"\nu = 5"];
```

```julia
#p = gr()
p = plot(
    rhos,
    vals,
    xlabel="t",
    ylabel=L"\bar{f}_t",
    #yformatter=y->(string(floor(Int,y*100))*"%"),
    label=[policies[1] policies[2] policies[3]],
    lw=7,
    linecolor = [:green :blue :purple],
    leg=(0.8,0.95),
    windowsize=(900,600),
    xtickfontsize = 11,
    ytickfontsize = 11,
    guidefont=font(30),
    bottom_margin=10mm,
    left_margin=10mm,
    frame=:box,
    legendfontsize = 14,
    widen = true,
    fmt = :png
        )
savefig(p, "./plots/sensitivity/procurement-avgAmount-case-study.png")
```
