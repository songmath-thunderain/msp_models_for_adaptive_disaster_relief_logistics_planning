```julia
using CSV, DataFrames, Plots, Plots.PlotMeasures, StatsPlots, LaTeXStrings, Statistics
```

```julia
rho = ["0.05", "0.5", "5"];
avg_data = Array{Float64, 2}(undef, 3, 4);
avg_data[1,:] = [78,112,128,141]
avg_data[2,:] = [122,134,122,134]
avg_data[3,:] = [85,85,85,85]
```

```julia
policies = ["FA-MSP","Wait-and-see","RH-2SSP","S-2SSP"];
```

```julia
p = groupedbar(
    rho,
    avg_data,
    xlabel=L"\nu",
    ylabel=L"\textrm{Gap}~(\%)",
    label=[policies[1] policies[2] policies[3] policies[4]],
    bar_position = :dodge,
    bar_width=0.7,
    color = [:red :yellow :lightgreen :blue],
    leg=(0.09,0.955),
    windowsize=(1000,550),
    xtickfontsize = 12,
    ytickfontsize = 12,
    guidefont=font(25),
    bottom_margin=10mm,
    left_margin=5mm,
    frame=:box,
    legendfontsize = 12,
    widen = true,
    legend=:topright,
    fmt = :png
    )
savefig(p, "./plots/averages/all_case_study_option0.png")
```
