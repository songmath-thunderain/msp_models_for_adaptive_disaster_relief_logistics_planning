```julia
using CSV, DataFrames, Plots, Plots.PlotMeasures, LaTeXStrings
```

```julia
states = [1, 3, 5]
count = 0;
myplots = [];
for s=1:3
    sensitivity = Matrix(CSV.read("sensitivity-2.csv",DataFrame));
    sensitivity = sensitivity[1+10*(s-1):10*s,:];
    print("sensitivity = ", sensitivity);
    costf = [0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5]
    T = collect(1:1:6)
    p1 = plot3d(1, 
        xlim = (1,6),
        ylim=(0,0.5),
        zlim = (0.0,maximum(sensitivity)+5),
        label = false)
    clcount=1.0;
    clcount2=0.0;
    for j=1:10
        x = [];
        y = [];
        z = [];
        for i=1:6
            push!(x, T[i])
            push!(y, costf[j])
            push!(z, sensitivity[j,i])
        end 
        plot!(x,y,z, label = false, lw=9, lc=RGB(clcount2,0,clcount))
        clcount -= 0.1
        clcount2 += 0.1
    end
    plot!(p1,
        xlabel=L"t",
        ylabel= L"\nu",
        #zlabel="average procured items",
        zlabel=L"\bar{f}_t",
        lw=5,
        windowsize=(800,500),
        xtickfontsize = 12,
        ytickfontsize = 12,
        ztickfontsize = 12,
        guidefont=25,
        bottom_margin=10mm,
        left_margin=10mm,
        #right_margin=12mm,
        gridlinewidth = 1,
        camera = (20,60),
        fmt = :png)
        push!(myplots,p1)
        savefig(p1, "../plots/sensitivity/p"*string(states[s])*".png")
end
```
